```python
ages = []

images = []

place = imdbPlace
where = 'imdb_crop'

total = 0
target_size = (100, 100)

for i in range(50000):
    if i % 10000 == 0:
        print(i)

    bYear = int(place[0][0][i] / 365)
    taken = place[1][0][i]
    path = place[2][0][i][0]
    gender = place[3][0][i]
    name = place[4][0][i]
    faceBox = place[5][0][i]
    faceScore = place[6][0][i]
    secFaceScore = place[7][0][i]
    age = taken - bYear
    faceScore = str(faceScore)
    secFaceScore = str(secFaceScore)
    if 'n' not in faceScore:
        if 'a' in secFaceScore:
            if age >= 0:
                try:
                    gender = int(gender)
                    total += 1

                    image_path = os.path.join(where, path)
                    print("Image path:", image_path)

                    image = Image.open(image_path)
                    resized_img = image.resize(target_size, Image.LANCZOS)
                    grayscale_img = resized_img.convert('L')
                    images.append(grayscale_img)
                    print(i)
                    ages.append(age)
                except Exception as e:
                    print('Failed with gender:', e)
                    continue
print("Total:", total)
```

```python
import cv2
import numpy as np


def augment_data(images_and_ages, num_copies):
    augmented_data = []
    for image, age in images_and_ages:
        augmented_data.append((image, age))

        mirrored_image = cv2.flip(image, 1)
        augmented_data.extend([(mirrored_image, age)] * num_copies)


        rotation_matrix_20 = cv2.getRotationMatrix2D((image.shape[1]/2, image.shape[0]/2), 20, 1)
        rotated_image_20 = cv2.warpAffine(image, rotation_matrix_20, (image.shape[1], image.shape[0]))
        augmented_data.extend([(rotated_image_20, age)] * num_copies)

        rotation_matrix_40 = cv2.getRotationMatrix2D((image.shape[1]/2, image.shape[0]/2), 40, 1)
        rotated_image_40 = cv2.warpAffine(image, rotation_matrix_40, (image.shape[1], image.shape[0]))
        augmented_data.extend([(rotated_image_40, age)] * num_copies)

        rotation_matrix_anticlock_20 = cv2.getRotationMatrix2D((image.shape[1]/2, image.shape[0]/2), -20, 1)
        rotated_image_anticlock_20 = cv2.warpAffine(image, rotation_matrix_anticlock_20, (image.shape[1], image.shape[0]))
        augmented_data.extend([(rotated_image_anticlock_20, age)] * num_copies)

        rotation_matrix_anticlock_40 = cv2.getRotationMatrix2D((image.shape[1]/2, image.shape[0]/2), -40, 1)
        rotated_image_anticlock_40 = cv2.warpAffine(image, rotation_matrix_anticlock_40, (image.shape[1], image.shape[0]))
        augmented_data.extend([(rotated_image_anticlock_40, age)] * num_copies)

    return augmented_data
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, ReLU, MaxPooling2D, GlobalAveragePooling2D, Dense, Dropout, Add
from tensorflow.keras.models import Model

def conv_block(inputs, filters, kernel_size, strides=(1, 1), padding='same'):
    x = Conv2D(filters, kernel_size, strides=strides, padding=padding)(inputs)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    return x

def identity_block(inputs, filters, kernel_size, strides=(1, 1), padding='same'):
    x = conv_block(inputs, filters, kernel_size, strides, padding)
    x = Conv2D(filters, kernel_size, strides=strides, padding=padding)(x)
    x = BatchNormalization()(x)
    x = Add()([inputs, x])
    x = ReLU()(x)
    return x

def ResNet18(input_shape=(100, 100, 1), num_classes=1):
    inputs = Input(shape=input_shape)

    x = conv_block(inputs, 64, (7, 7), strides=(2, 2))
    x = MaxPooling2D((3, 3), strides=(2, 2), padding='same')(x)

    x = identity_block(x, 64, (3, 3))
    x = identity_block(x, 64, (3, 3))

    x = conv_block(x, 128, (3, 3), strides=(2, 2))
    x = identity_block(x, 128, (3, 3))
    x = identity_block(x, 128, (3, 3))

    x = conv_block(x, 256, (3, 3), strides=(2, 2))
    x = identity_block(x, 256, (3, 3))
    x = identity_block(x, 256, (3, 3))

    x = GlobalAveragePooling2D()(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='linear', name='age_output')(x)

    model = Model(inputs, outputs, name='resnet18')
    return model

```
